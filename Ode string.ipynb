This time we want to solve for the position of a wave pulse on a weighted string with tension:

$$T(x) = \sqrt{(1 - x/l)g + Mg}$$

$$V(x) = \sqrt{(l - x)g + Mgl / m}$$

```python
%matplotlib inline
from __future__ import division
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt
from math import sqrt

g = 10.0 #ms^-1
l = 7.0 #m
m = 1 #Kg
M = m / 9

def Y_dash_pulse(Y, t):
    return sqrt((l - Y)*g + M*g*l / m) if l - Y > 0 else 0

def Y_dash_falling_body(Y, t):
    return [Y[1], g]

X = np.arange(0, sqrt(2*l / g) * 1.1, 0.01)
pulse = integrate.odeint(Y_dash_pulse, 0, X)
body = integrate.odeint(Y_dash_falling_body, [0, 0], X)
plt.plot(X, body[:, 0])
plt.plot(X, pulse)
plt.plot([0, 1.4], [l, l])
plt.annotate('intersection', xy=(sqrt(2*l / g), l), xytext=(sqrt(2*l / g) -0.5, l +1),
            arrowprops=dict(facecolor='black', shrink=0.08),
            )
plt.show()
```

```python
from __future__ import division
import pylab
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from cmath import exp, pi, phase
from math import cos, sin, sqrt
from scipy.integrate import quad, fixed_quad
plt.rcParams['figure.figsize'] = (20.0, 15.0)

def phi(Y, D):
    wavelength = 500e-9
    d = 1e-3
    k = 2*pi/wavelength
    
    def real_wave(y): return cos(k * sqrt((Y - y)**2 + D**2))
    def imag_wave(y): return sin(k * sqrt((Y - y)**2 + D**2))
    
    real_part = quad(real_wave, -d/2, d/2)[0]
    imag_part = quad(imag_wave, -d/2, d/2)[0]
    return real_part**2 + imag_part**2

Ds = [5, 10, 100, 326, 640, 4000]
f, axes = plt.subplots(len(Ds))
for n,D in enumerate(Ds):
    X = np.linspace(-0.005*D, 0.005*D, 1000)
    axes[n].set_title("D = {}mm".format(D))
    Y = [phi(x, D * 1e-3) for x in X]
    axes[n].plot(X,Y)
    

```

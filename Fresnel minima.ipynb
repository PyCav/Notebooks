```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
from scipy.special import fresnel, fresnel_zeros
from scipy.optimize import minimize
from mpmath import fresnels, fresnelc

x = np.linspace(0, 10, 1000)
def A(x):
    f_sin, f_cos = fresnel(x)
    return np.sqrt(f_sin**2 + f_cos**2)

@np.vectorize
def A2(x):
    return np.sqrt(fresnels(x)**2 + fresnelc(x)**2)


f, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, A(x))
minima_guesses = (1.9,2.75, 3.4, 4.12)
minima = np.array([minimize(A2, x0 = guess, tol = 0.000001).x for guess in minima_guesses])
ax1.scatter(minima_guesses, [0.50] * len(minima_guesses))
ax1.scatter(minima, [0.55] * len(minima_guesses))

n = range(len(minima))
ax2.scatter(n,minima)
```

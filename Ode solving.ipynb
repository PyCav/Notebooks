Before we can solve the ode we need to rewrite it as 2 coupled 1st order equations so

$$\ddot{\theta} = -\frac{g}{l} \; \sin{\theta}$$

becomes

$$\frac{d}{dx}\begin{bmatrix} y_0 \\ y_1 \end{bmatrix} = \begin{bmatrix}
    y_1 \\
    -\frac{g}{l} \; \sin{y_0}
  \end{bmatrix}$$

```python
%matplotlib inline
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt
from math import sin

g = 10.0 #ms^-1
l = 10.0 #m

def Y_dash(Y, t):
    return [Y[1], -g/l * sin(Y[0])]

X = np.arange(0, 25.0, 0.01)
for initial_val in [0.01, 0.03, 0.1]:
    out = integrate.odeint(Y_dash, [initial_val, 0], X)
    plt.plot(X, out[:, 0])
plt.show()
```

For low amplitudes the waveform is sinusoidal and the period is amplitude independant.

```python
for initial_val in [1.0, 2.0, 3.0]:
    out = integrate.odeint(Y_dash, [initial_val, 0], X)
    plt.plot(X, out[:, 0])
plt.show()
```

For higher amplitudes neither property holds.
